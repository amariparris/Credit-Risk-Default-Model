```python
# Coerce DateTime
eco.YEAR = pd.to_datetime(eco['YEAR'])
train['issueDate'] = pd.to_datetime(train['issueDate'])
```

```python
# Make endDate Column
train['endDate'] = train.apply(lambda x: x['issueDate'] + pd.DateOffset(years=int(x['term'])), axis=1)
```

```python
# Init the avg columns
train['avg_GDP'] = 0
train['avg_UNRATE'] = 0
train['avg_CPIAUCSL'] = 0
```

```python
# Function to Generate the averages - very inefficient.
def calculate_average_metrics(train_df, eco_df):
    
    # Iterate over each loan in train_df
    for index, loan in train_df.iterrows():
        
        # Identify the quarters within the loan period
        relevant_quarters = eco_df[(eco_df['YEAR'] >= loan['issueDate']) & (eco_df['YEAR'] <= loan['endDate'])]

        # Calculate averages if relevant_quarters is not empty
        if not relevant_quarters.empty:
            train_df.at[index, 'avg_GDP'] = relevant_quarters['GDP'].mean()
            train_df.at[index, 'avg_UNRATE'] = relevant_quarters['UNRATE'].mean()
            train_df.at[index, 'avg_CPIAUCSL'] = relevant_quarters['CPIAUCSL'].mean()

    return train_df
```

```python
# check that approrpiate columns returned
for index, loan in train.head(2)[:1].iterrows():
    s = eco[(eco['YEAR'] >= loan['issueDate']) & (eco['YEAR'] <= loan['endDate'])]
    print(s)
    
```

```python
# Call the Function
calculate_average_metrics(train,eco)
```

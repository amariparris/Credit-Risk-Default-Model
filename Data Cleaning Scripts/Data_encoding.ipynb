```python
# Assign numericals to subgrades 

def grade_to_numeric(input_value, input_type='subgrade'):
    # base value for each grade
    base_values = {'A': 6, 'B': 5, 'C': 4, 'D': 3, 'E': 2, 'F': 1, 'G': 0}
    
    if input_type == 'subgrade':
        # Extract the grade letter and numeric part from the subgrade
        grade_letter = input_value[0]
        numeric_part = int(input_value[1])
        
        # Calculate the decrement factor to adjust values within a grade
        decrement_factor = 0.5 / 5
        
        # Calculate the numerical value for subgrades
        numeric_value = base_values[grade_letter] - decrement_factor * (6 - numeric_part)
        
        # Adjust the numeric_value to match the provided scale starting from 0.4
        adjustment = 0.4 - (base_values['G'] - decrement_factor * 5)
        numeric_value += adjustment
        
    elif input_type == 'grade':

        grade_letter = input_value
        numeric_value = base_values[grade_letter]
        
    return round(numeric_value, 1)


# Apply the function to the subGrade column
data['num_subGrade'] = data['subGrade'].apply(lambda x: grade_to_numeric(x, 'subgrade'))

# Apply the function to the 'grade' column
data['num_grade'] = data['grade'].apply(lambda x: grade_to_numeric(x, 'grade'))
```

```python
# Extract the numeric part of the 'employmentLength' entries
data['employmentLength'] = data['employmentLength'].str.extract('(\d+)')

# Convert the extracted string numbers to integers
data['employmentLength'] = pd.to_numeric(data['employmentLength'], errors='coerce')
```
